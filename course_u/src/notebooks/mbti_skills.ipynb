```python
import pandas as pd
import numpy as np
```

```python
mbti_skills = pd.read_csv('../../static/csv/mbti_skills.csv')
mbti_skills.head()
```

```python
website_skill = pd.read_csv('../../static/csv/skill_2.csv')
website_skill.head()
```

```python
# convert skill_list to list
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: x.split(','))
mbti_skills.head()
```

```python
# remove [ ] from skill_list
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [i.replace('[', '').replace(']', '') for i in x])
```

```python
mbti_skills.head()
```

```python
# make  skill_list column lower case
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [i.lower() for i in x])
mbti_skills.head()
```

```python
# save to csv
mbti_skills.to_csv('../../static/csv/mbti_skills.csv', index=False)
```

```python
# Create a list of skills by combining all the skills from the 'skills_list' column
all_skills = [skill for skill_list in mbti_skills['skill_list'] for skill in skill_list]

# Now, 'all_skills' contains all the skills from the DataFrame
print(all_skills)
```

```python
# make it lower case
all_skills = [skill.lower() for skill in all_skills]
```

```python
# from mbti skill_list, combine all skill_list list into one list
mbti_skill_list = []
for skill_list in mbti_skills['skill_list']:
    for skill in skill_list:
        mbti_skill_list.append(skill)

mbti_skill_list
```

```python
# remove space at front and back of skill
mbti_skill_list = [skill.strip() for skill in mbti_skill_list]
mbti_skill_list
```

```python
# convert to small letter
mbti_skill_list = [skill.lower() for skill in mbti_skill_list]

# on mbti_skill, convert also to small letter
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [i.lower() for i in x])
mbti_skills.head()
```

```python
# on website_skill, contain the skill into a list
# get the skill value of all rows in the 'skill' column
all__existing_skills = website_skill['skill'].to_list()
all__existing_skills
```

```python
# on mbti_skill_list, remove the skill that already exist in all__existing_skills
mbti_skill_list = [skill for skill in mbti_skill_list if skill not in all__existing_skills]
mbti_skill_list
```

```python
print(len(mbti_skill_list))
```

```python
# # convert the list of unique skills to a DataFrame
skills_df = pd.DataFrame(mbti_skill_list, columns=['skill'])
skills_df.head()
```

```python
# get last id of skill
last_id = website_skill['id'].max()
last_id
```

```python
# add id to skills_df
skills_df['id'] = skills_df.index + last_id + 1
# reorder
skills_df = skills_df[['id', 'skill']]
skills_df.head()
```

```python
# combine skills_df and website_skill
website_skill = pd.concat([website_skill, skills_df])
website_skill.tail()
```

```python
# save to csv
website_skill.to_csv('../../static/csv/skill_3.csv', index=False)
```
