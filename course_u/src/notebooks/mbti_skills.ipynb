# MBTI Skill Mapping

```python
import numpy as np
import pandas as pd
import ast
```

```python
# load skill_3.csv
website_skill = pd.read_csv('../../static/csv/skill_3.csv')
website_skill.head(3)
```

```python
# load mbti_skills.csv
mbti_skills = pd.read_csv('../../static/csv/mbti_skills.csv')
mbti_skills.head(3)
```

```python
# use ast to convert string to list
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: ast.literal_eval(x))
```

```python
# convert skill_list to mbti_skill_list
mbti_skill_list = []
for i in mbti_skills['skill_list']:
    mbti_skill_list += i

mbti_skill_list
```

```python
# strip space at front and back of skill
mbti_skill_list = [skill.strip() for skill in mbti_skill_list]
mbti_skill_list
```

```python
# convert website_skill to dictionary
# skill_mapping = unique_skills_df.set_index('skill')['id'].to_dict()
website_skill_dict = website_skill.set_index('skill')['id'].to_dict()
```

```python
# create dictionary of skill and id
website_skill_dict
```

```python
mbti_skills.head(3)
```

```python
# strip space at front and back of skill
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [skill.strip() for skill in x])
```

```python
# remove '-' and ' ' from skill
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [skill.replace('-', ' ') for skill in x])
```

```python
# remve 'skills' and 'skill' from skill
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [skill.replace('skills', '').replace('skill', '') for skill in x])
```

```python
# remove empty string
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [skill for skill in x if skill != ''])
# strip
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [skill.strip() for skill in x])
```

```python
# apply skill_mapping to skill_list
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [website_skill_dict[i] for i in x])
mbti_skills.head()
```

```python
# explode
mbti_skills = mbti_skills.explode('skill_list')
# get id, and skill_list
mbti_skills = mbti_skills[['id', 'skill_list']]
mbti_skills.head()
```

```python
# rename column
mbti_skills.rename(columns={'skill_list': 'skill_id', 'id':'indicator_id'}, inplace=True)
mbti_skills.sample(2)
```

```python
# reset index
mbti_skills.reset_index(drop=True, inplace=True)
```

```python
# add id
mbti_skills['id'] = mbti_skills.index + 1
mbti_skills.head(3)
```

```python
mbti_skills.tail(3)
```

```python
# reoder column
mbti_skills = mbti_skills[['id', 'indicator_id', 'skill_id']]
mbti_skills.head(3)
```

```python
# save to csv
mbti_skills.to_csv('../../static/csv/personality_indicator_skills.csv', index=False)
```
