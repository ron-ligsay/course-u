```python
# on website_skill, contain the skill into a list
# get the skill value of all rows in the 'skill' column
website_skill_list = website_skill['skill'].to_list()
website_skill_list
```

```python
# on mbti_skill_list, remove the skill that already exist in all__existing_skills
mbti_unique_skill_list = [skill for skill in mbti_skills_list if skill not in website_skill_list]
mbti_unique_skill_list
```

```python
print(len(mbti_unique_skill_list))
```

```python
# # convert the list of unique skills to a DataFrame
skills_df = pd.DataFrame(mbti_unique_skill_list, columns=['skill'])
skills_df.head()
```

```python
# get last id of skill
last_id = website_skill['id'].max()
last_id
```

```python
# add id to skills_df
skills_df['id'] = skills_df.index + last_id + 1
# reorder
skills_df = skills_df[['id', 'skill']]
skills_df.head()
```

```python
website_skill.tail(3)
```

```python
# combine skills_df and website_skill
website_skill = pd.concat([website_skill, skills_df])
website_skill.tail()
```

```python
# reset index
website_skill.reset_index(drop=True, inplace=True)
website_skill.tail()
```
