```python
import pandas as pd
```

```python
website_skill = pd.read_csv('../../static/csv/skill_3.csv')
website_skill.tail()
```

```python
skill_mapping = website_skill.set_index('skill')['id'].to_dict()
skill_mapping
```

```python
# load csv test_
test_ = pd.read_csv('../../static/csv/test_.csv')
test_.tail()
```

```python
# use ast to convert string to list
import ast
test_['skills_list'] = test_['skills_list'].apply(lambda x: ast.literal_eval(x))
test_.tail()
```

```python
# apply skill mapping
test_['skill_id'] = test_['skills_list'].apply(lambda x: [skill_mapping[i] for i in x])
test_.tail()
```

```python
# explode skill_id
test_ = test_.explode('skill_id')
test_ = test_[['question_id', 'skill_id']]
test_.tail()
```

```python
# add id, starting to one
test_['id'] = test_.index + 1
test_.tail()
```

```python
# rename question_id to test_id
test_ = test_.rename(columns={'question_id': 'test_id'})
# reorder columns
test_ = test_[['id', 'test_id', 'skill_id']]
test_.tail()
```

```python
# save to csv
test_.to_csv('../../static/csv/assessment_test_skill_.csv', index=False)
```

```python
# laod mbti_skills
mbti_skills = pd.read_csv('../../static/csv/mbti_skills.csv')
mbti_skills.tail()
```

```python
# use ast to convert string to list
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: ast.literal_eval(x))
mbti_skills.tail()
```

```python
# for each row, on each skill in list remove space in front and back
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [i.strip() for i in x])
```

```python
# apply skill mapping
mbti_skills['skill_id'] = mbti_skills['skill_list'].apply(lambda x: [skill_mapping[i] for i in x])
mbti_skills.tail()
```

```python
# explode skill_id
mbti_skills = mbti_skills.explode('skill_id')
mbti_skills = mbti_skills[['id', 'skill_id']]
# rename id to mbti_id
mbti_skills = mbti_skills.rename(columns={'id': 'mbti_id'})
mbti_skills.tail()
```

```python
# add id, starting to one
mbti_skills['id'] = mbti_skills.index + 1
# reorder columns
mbti_skills = mbti_skills[['id', 'mbti_id', 'skill_id']]
mbti_skills.tail()
```

```python
# save to csv
mbti_skills.to_csv('../../static/csv/personality_mbti_skill_.csv', index=False)
```
