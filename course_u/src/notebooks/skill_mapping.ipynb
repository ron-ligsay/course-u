```python
# laod mbti_skills
mbti_skills = pd.read_csv('../../static/csv/indicator_.csv')
mbti_skills.head()
```

```python
mbti_skills.head()
```

```python
# on skill_list with list of strings replace '-' with ' '
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [i.replace('-', ' ') for i in x])
```

```python
# lower case
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [i.lower() for i in x])
mbti_skills.tail()
```

```python
# check skill list type if its a list
mbti_skills['skill_list'].apply(lambda x: type(x))
```

```python
mbti_skills.head()
```

```python
# Use ast.literal_eval to convert the skill_list strings to lists
for i in range(len(mbti_skills['skill_list'])):
    mbti_skills['skill_list'][i] = ast.literal_eval(mbti_skills['skill_list'][i])

mbti_skills.head()
```

```python
# for each row, on each skill in list remove space in front and back
mbti_skills['skill_list'] = mbti_skills['skill_list'].apply(lambda x: [i.strip() for i in x])
```

```python
mbti_skills.head()
```

```python
# apply skill mapping
mbti_skills['skill_id'] = mbti_skills['skill_list'].apply(lambda x: [skill_mapping[i] for i in x])
mbti_skills.tail()
```

```python
# explode skill_id
mbti_skills = mbti_skills.explode('skill_id')
mbti_skills = mbti_skills[['indicator_id', 'skill_id']]
# rename id to mbti_id
#mbti_skills = mbti_skills.rename(columns={'indicator_id': 'indicator_id'})
mbti_skills.tail()
```

```python
# reset index
mbti_skills = mbti_skills.reset_index(drop=True)
```

```python
# add id, starting to one
mbti_skills['id'] = mbti_skills.index + 1
# reorder columns
mbti_skills = mbti_skills[['id', 'indicator_id', 'skill_id']]
mbti_skills.tail()
```

```python
# save to csv
mbti_skills.to_csv('../../static/csv/personality_indicator_skills.csv', index=False)
```
