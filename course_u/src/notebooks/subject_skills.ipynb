```python
import ast
# Convert the string representation of lists into actual lists
acad_subject['skill_list'] = acad_subject['skill_list'].apply(ast.literal_eval)
acad_subject.head()
```

```python
# on the column skill_list make it to lower case
acad_subject['skill_list'] = acad_subject['skill_list'].apply(lambda x: [y.lower() for y in x])
acad_subject.head()
```

```python
# replace '-' with ' ' in the column skill_list
acad_subject['skill_list'] = acad_subject['skill_list'].apply(lambda x: [y.replace('-', ' ') for y in x])
```

```python
# remove 'skills' and 'skill'
acad_subject['skill_list'] = acad_subject['skill_list'].apply(lambda x: [y.replace('skills', '') for y in x])
```

```python
# save to csv
acad_subject.to_csv('../../static/csv/subject_.csv', index=False)
```

```python
# Create a list of skills by combining all the skills from the 'skills_list' column
subject_skills = [skill for skill_list in acad_subject['skill_list'] for skill in skill_list]

# Now, 'all_skills' contains all the skills from the DataFrame
print(subject_skills)
```

```python
# get unique skills
unique_subject_skills = list(set(subject_skills))
print(unique_subject_skills)
# count the number of unique skills
print(len(unique_subject_skills))
```
