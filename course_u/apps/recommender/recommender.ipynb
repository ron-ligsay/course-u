```python
import pandas as pd
import numpy as np
import ast
```

```python
# load csv
specialization = pd.read_csv('../../static/csv/specialization3_skills.csv')
specialization.head()
```

```python
# load csv
website_skill = pd.read_csv('../../static/csv/skill_4.csv')
website_skill.head()
```

```python
# apply ast to convert string to list
specialization['skills'] = specialization['skill_list'].apply(ast.literal_eval)
specialization.sample(3)
```

```python
# # Group by 'specialization' and aggregate skills with their frequencies
# specialization = specialization.groupby('title')['skills'].sum().reset_index()
# # fill na with 0
# specialization['skills'] = specialization['skills'].fillna(0)
# specialization.sample(3)
```

```python
# remove empty '' in skills
specialization['skills'] = specialization['skills'].apply(lambda x: [i for i in x if i != ''])
```

```python
# # aggregate skills with their frequencies
# specialization = specialization.groupby('title')['skills'].sum().reset_index()
# # fill na with 0
# specialization['skills'] = specialization['skills'].fillna(0)
# specialization.sample(3)
```

```python
specialization.sample(3)
```

```python
# Get all the unique skills.
unique_skills = []
for i in specialization['skills']:
    unique_skills += i
```

```python
# get website skills
website_skill_list = website_skill['skill'].tolist()
```

```python
# add website skills to unique skills
unique_skills += website_skill_list
```

```python
# remove duplicates
unique_skills = list(set(unique_skills))
```

```python
# remove blank ''
unique_skills = [i for i in unique_skills if i != '']
```

```python
# remove ' ' 
unique_skills = [i for i in unique_skills if i != ' ']
```

```python
# get samples
unique_skills[:5]
```

```python
# Create a new data frame with the unique skills as columns.
specialization_sparse = pd.DataFrame(columns=['title'] + list(unique_skills))
specialization_sparse.head()
```

```python
# Iterate over the specialization data frame and fill in the new data frame.
for i in range(len(specialization)):
    title = specialization.loc[i, 'title']
    skills = specialization.loc[i, 'skills']

    # Create a dictionary mapping skill names to their frequencies.
    skill_counts = {}
    for skill in skills:
        if skill not in skill_counts:
            skill_counts[skill] = 0
        skill_counts[skill] += 1

    # Fill in the new data frame.
    specialization_sparse.loc[i, 'title'] = title
    for skill in unique_skills:
        if skill in skill_counts:
            specialization_sparse.loc[i, skill] = skill_counts[skill]
        else:
            specialization_sparse.loc[i, skill] = 0
```

```python
specialization_sparse.sample(10)
```

```python
# Calculate total skills for each role
specialization_sparse['total_skills'] = specialization_sparse.drop('title', axis=1).sum(axis=1)
specialization_sparse.head()
```

```python
# Normalize skill frequencies by dividing by total skills
specialization_sparse.iloc[:, 1:-1] = specialization_sparse.iloc[:, 1:-1].div(specialization_sparse.total_skills, axis=0)
specialization_sparse.head()
```

```python
specialization_sparse.sample(3)
```

```python
# from specialization include field_id to specialization_sparse
specialization_sparse = pd.merge(specialization_sparse, specialization[['title', 'field_id']], on='title')
```

```python
specialization_sparse
```

```python
# move field_id column to the second column
cols = specialization_sparse.columns.tolist()
cols = cols[-1:] + cols[:-1]
specialization_sparse = specialization_sparse[cols]
specialization_sparse.head()
```

```python
# save to csv
specialization_sparse.to_csv('specialization_sparse.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Group the data by field_id and skill_id, and calculate the sum of the values
grouped = field_skill.groupby(['field_id', 'skill_id']).value.sum().reset_index()

# Get the unique field_ids
field_ids = grouped.field_id.unique()

# For each field_id
for field_id in field_ids:
    # Get the data for this field_id
    data = grouped[grouped.field_id == field_id]
    
    # Sort the data by value and get the top 12 skills
    top_skills = data.sort_values(by='value', ascending=False).head(12)
    
    # Create a bar plot
    plt.figure(figsize=(10, 6))
    plt.bar(top_skills.skill_id, top_skills.value)
    plt.title(f'Field ID: {field_id}')
    plt.xlabel('Skill ID')
    plt.ylabel('Value')
    plt.show()
```
