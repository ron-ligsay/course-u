```python
import pandas as pd
```

```python
jobs = pd.read_csv('jobs_2.csv')
jobs.head()
```

```python
# count keyword
jobs['keyword'].value_counts()
```

```python
# display keyword if equal  to keyword
#jobs[jobs['keyword'] == 'keyword']
# drop keyword if equal to keyword
jobs = jobs[jobs['keyword'] != 'keyword']
```

```python
def remove_query_parameters(url):
    if '&' in url:
        return url.split('&')[0]
    return url
```

```python
jobs['link_clean'] = jobs['link'].apply(remove_query_parameters)
jobs.head()
```

```python
# Add ID column
jobs['id'] = jobs.index + 1
jobs.head()
```

```python
# reorder columns
jobs = jobs[['id', 'keyword', 'title', 'company', 'link', 'company_link', 'date']]
jobs.head()
```
